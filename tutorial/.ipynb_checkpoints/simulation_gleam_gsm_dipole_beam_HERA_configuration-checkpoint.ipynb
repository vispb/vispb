### This notebook demonstrates how to run a simple visibility simulation with GLEAM and GSM2016, and a dipole beam

```python
import numpy as np
import matplotlib.pyplot as plt
import os

from VisPB import Sky_Model, Beam_Model, Vis_Simulator
from VisPB.data import DATA_PATH
```

### Setting up the antenna configuration

Antenna configuration should be set up in a dictionary form for the visibility simulation. We use actual HERA configuration for the first 60 antennas.

```python
ant_filename = os.path.join(DATA_PATH, 'hera_antenna_positions_350.dat')

ant_config = np.loadtxt(ant_filename, delimiter=',')
ants_layout = np.array(ant_config[:,0], dtype=int)
array_layout = ant_config[:,1:]
ants_dict = {ant: array_layout[i, :] for i, ant in enumerate(ants_layout)}
```

```python
ants = np.arange(60)
antenna_config = {ant: ants_dict[ant] for ant in ants}
```

```python
ant_pos = np.asarray(list(antenna_config.values()))
x = ant_pos[:,0]
y = ant_pos[:,1]
buffer_plot = 10

plt.figure(figsize=(3,3), dpi=150)
plt.scatter(x, y, s=80)
for i, (x_i, y_i) in enumerate(zip(x, y)):
    plt.text(x_i, y_i, i, horizontalalignment='center', verticalalignment='center', fontsize=6)
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.xlim(x.min()-buffer_plot, x.max()+buffer_plot)
plt.ylim(y.min()-buffer_plot, y.max()+buffer_plot)
```

### Constructing and reading sky model using pyradiosky

Sky_Model can read "gleam", "gsm08", "gsm16", and pyradiosky-readable dictionary or file.

Here we give an example for GLEAM + GSM2016.

```python
Nfreqs = 121
freqs = np.linspace(160e6, 180e6, Nfreqs) # frequencies where the flux density will be evaluated
```

```python
sky_model = Sky_Model(["gleam", "gsm16"], freqs, nside=64)
sky_model.set_flux_cut("gleam", 160e6, 1) # selecting gleam point sources brighter than 1 Jy at 160 MHz
```

```python
print('sky model:', np.unique(sky_model.component))
```

### Setting up the beam object for the cst-simulated dipole beam

Analytic beams (uniform, gaussian, and airy disk) or cst-simulated beams can be read.

```python
beam_filename = "file path to your beam file"
beam_model = Beam_Model(freqs)
beam_model.read_cst_beam(0, ants, beam_filename, set_x_orientation='EAST', efield_to_power=False)
```

```python
beam_model.beam_dict
```

### Setting up and starting the visibility simulation

Sky_Model and Beam_Model defined above will be used in Vis_Simulator.

```python
sim = Vis_Simulator(sky_model, beam_model, antenna_config,
                    ants=ants, pols=['xx', 'xy'], memory_limit=2, Nthread=8)
```

```python
time_array = 2458209.93985
sim.run(time_array, integration_time=10)
```

You can convert the Vis_Simulator object to UVData object for each sky model or combined sky model.

```python
uvd_gleam = sim.get_UVData("gleam")
uvd_gsm = sim.get_UVData("gsm16")
uvd_fg = sim.get_UVData(["gleam", "gsm16"]) # combining two sky models
```

Redundancy check for redundant baselines

```python
bl = (0, 1)
pols = ['ee', 'en']

lsts_hr = np.rad2deg(np.unique(uvd_fg.lst_array)) / 15
freqs_MHz = freqs * 1e-6

fig = plt.figure(figsize=(9,4), dpi=100)
axes = fig.subplots(1,2)
for i, pol in enumerate(pols):
    ax = axes[i]
    key = bl+(pol,)
    ax.plot(freqs_MHz, np.abs(uvd_gleam.get_data(key)[0]), label='GLEAM')
    ax.plot(freqs_MHz, np.abs(uvd_gsm.get_data(key)[0]), label='GSM')
    ax.plot(freqs_MHz, np.abs(uvd_fg.get_data(key)[0]), label='GLEAM+GSM')
    ax.set_title(key)
    ax.set_xlabel('Frequency (MHz)')
    if(i == 0):
        ax.set_ylabel('LST (hr)')
    ax.legend()

fig.subplots_adjust(wspace=0.25)
```
